```python
kern_win_size = 21  # it was 2001 
x_kern = sp.linspace(-6,6,kern_win_size)
kern = sp.exp(-x_kern**2/4)
kern /= kern.sum() # equivalent to kern = kern/kern.sum()

# GDD_avg = sp.convolve(Gdd,kern,'same')
GDD_avg1 = sp.convolve(Gdd1,kern,'same')
plt.figure()
# plt.plot(wl*1e9,GDD_avg)
plt.plot(wl1,GDD_avg1)
plt.xlabel('Wavelength(nm)')
plt.ylabel('Gdd (fs^2)')
plt.xlim(1026,1120)
plt.ylim(-500,500)

plt.savefig('GDD_250fs', format = None)
```
